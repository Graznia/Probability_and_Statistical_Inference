# Histogram AGES

```r
AGES<-c(18, 18, 18, 18, 19, 19, 20, 20, 58)

hist(AGES,100)
```

```r
#  MEAN
mean(AGES)
# MEDIAN
median(AGES)
# STANDARD DEVIATION
sd(AGES)
# VARIANCE
var(AGES)
#RANGE
range(AGES)
# INTERQUARTILE RANGE
IQR(AGES)
# ANOTHER WAY OF GETTING THE INTERQUARTILE RANGE
Q1<-quantile(AGES,1/4)
Q3<-quantile(AGES,3/4)
INTERQRANGE<-Q3-Q1
```

# STANDARD DEVIATION


```r
x <- seq(-4, 4, length=100)
sd<-c(1,2)
hx <- dnorm(x,0,sd)

colors <- c("red", "blue" ,"yellow")
labels <- c("sd=1", "sd=2")

 plot(x, 0*x, type="l", lty=2, xlab="x value",
      ylab="Density",ylim=c(0, 1))

for (i in 1:2){
  lines(x, dnorm(x,0,sd[i]), lwd=2, col=colors[i])
}

legend("topright", inset=.05, title="Distributions",
       labels, lwd=2, lty=c(1, 1), col=colors)
```

```r
x <- seq(0,30,length=1000)
Rand_dist<-rchisq(10000,3)
x_mean<-c(mean(Rand_dist),mean(Rand_dist))
x_median<-c(median(Rand_dist),median(Rand_dist))
skew<-(mean(Rand_dist)-median(Rand_dist))/sd(Rand_dist)

hist_info<-hist(Rand_dist,100,main = paste("Histogram of Data"))
mode <- hist_info$mids[hist_info$counts == max(hist_info$counts)]
x_mode=c(mode,mode)
y<-c(0,max(hist_info$counts))

lines(x_mean,y, col="red")
lines(x_median,y, col="blue")
lines(x_mode,y, col="yellow")

colors <- c("red", "blue" ,"yellow")
labels <- c("mean","median","mode")

legend("topright", inset=.05, title="Distributions",
       labels, lwd=3, lty=c(1, 1), col=colors)
```
