# Geometric Mass Function

```r
x<- 0:10
par(mfrow=c(1,2))
p<-0.1
plot(x+1,dgeom(x,p), xlab="Number of Coins Tosses",ylab="P(X=x)",type="h",ylim=c(0,1))
plot(x+1,pgeom(x,p), xlab="Number of Coin Tosses",ylab="P(X<=x)",type="h", main="Cummulative First Head",ylim=c(0,1))
```

# Binomial Mass Distribution

```r
x<- 0:10
par(mfrow=c(1,2))
p<-0.5
plot(x,dbinom(x, size=10, prob=p), xlab="Number of Heads",ylab="P(X=x)",type="h",ylim=c(0,1))
plot(x,pbinom(x, size=10, prob=p), xlab="Number of Heads",ylab="P(X<=x)",type="h", main="Cummulative",ylim=c(0,1))
```
